```python
# --- Standard scientific libraries ---
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from mpl_toolkits.mplot3d import Axes3D  # for 3D plotting


# --- Reservoir implementations ---
from reservoirs.cycle import CycleReservoir3D
from reservoirs.sparse import SparseESN3D
from reservoirs.bei_stp import BEISTPReservoirESN
from reservoirs.swirl_gated_multicycle import SwirlGatedMultiCycleESN
from reservoirs.resonator import ResonatorReservoirESN
from reservoirs.neuromodulatory import NGCESN

# --- Utility modules ---
from utils.lorenz import generate_lorenz_data
from utils.metrics import compute_valid_prediction_time, compute_attractor_deviation, evaluate_nrmse, mse_dimwise
from utils.plotting import plot_timeseries, plot_lorenz_3d
from utils.dynamics import plot_eigen_spectrum
```

```python
# Set Seaborn theme
sns.set_theme(style="white")

# 1) Generate Lorenz
tmax = 250.0
dt   = 0.02
t_vals, lorenz_traj = generate_lorenz_data(
    initial_state=[1.0, 1.0, 1.0],
    tmax=tmax,
    dt=dt,
    sigma=10.0,
    rho=28.0,
    beta=8.0/3.0
)

washout = 2000
t_vals = t_vals[washout:]
lorenz_traj = lorenz_traj[washout:]

T_data = len(lorenz_traj)
print(f"Data length: {T_data}, from t=0..{tmax} with dt={dt}.")

train_frac = 0.8
train_end = int(train_frac*(T_data-1))
train_input  = lorenz_traj[:train_end]
train_target = lorenz_traj[1:train_end+1]
test_input   = lorenz_traj[train_end:-1]
test_target  = lorenz_traj[train_end+1:]
print(f"Train size: {len(train_input)}  Test size: {len(test_input)}")

# Create a time array (for VPT calculation)
n_test_steps = len(test_target)
time_test = np.arange(n_test_steps) * dt

# 2a) Train baseline models
cycle_res = CycleReservoir3D(
    reservoir_size=300, 
    spectral_radius=0.95,
    input_scale=0.2, 
    leaking_rate=0.8, 
    ridge_alpha=1e-4, 
    seed=43
)
cycle_res.fit_readout(train_input, train_target, discard=100)

sparse_res = SparseESN3D(
    reservoir_size=300, 
    spectral_radius=0.95,
    connectivity=0.05, 
    input_scale=0.2, 
    leaking_rate=0.8,
    ridge_alpha=1e-4, 
    seed=44
)
sparse_res.fit_readout(train_input, train_target, discard=100)

# 2b) Train proposed models
'''
sph_res = BEISTPReservoirESN(
    reservoir_size   = 300,
    sub_reservoirs   = 4,
    inter_density    = 0.01,
    inter_scale      = 0.2,
    e_frac           = 0.8,
    inhib_gain       = 1.5,
    spectral_radius  = 0.95,
    leak_rate        = 0.2,
    input_scale      = 0.1,
    U                = 0.10,
    tau_rec          = 500.0,
    tau_fac          = 100.0,
    ridge_alpha      = 1e-7,
    seed             = 123,
)

sph_res = SwirlGatedMultiCycleESN(
        reservoir_size       = 300,   # total neurons  (must be divisible by n_cycles)
        n_cycles             = 10,      # how many simple cycles
        cycle_weight         = 0.97,   # weight on each ring edge
        bridge_weight        = 0.20,   # weight on each sparse inter-cycle bridge
        input_scale          = 0.70,   # scaling factor ‖W_in‖
        leak_rate            = 0.20,   # α  (1.0 ⇒ no leak)
        ridge_alpha          = 1e-6,   # ℓ₂ penalty for ridge read-out
        swirl_beta           = 5.0,    # β  steepness of swirl gate
        swirl_frequency      = None,   # ω  (defaults to 2π / m if None)
        swirl_sigmoid        = True,   # True ⇒ σ(·) ∘ sin; False ⇒ raw sin
        seed                 = 42,     # RNG seed for reproducibility
        use_polynomial_readout = True  # include quadratic features in read-out
    )
'''

sph_res = ResonatorReservoirESN(
        reservoir_size      = 300,     # total neurons (must be even)
        frequency_mode      = "gold",   # {"log", "lin", "gold"}  – angle generator
        r_damp              = 0.96,    # intra-pair damping factor r
        eps_couple          = 0.03,    # cross-pair coupling strength ε
        input_scale         = 0.20,    # scaling of random W_in
        leak_rate           = 0.50,    # α  (1.0 ⇒ no leak)
        ridge_alpha         = 1e-6,    # ℓ₂ penalty for ridge read-out
        use_gain            = True,    # apply static per-neuron gains g_k ?
        gain_beta           = 0.5,     # β  in gain = σ[β·sin(·)]
        gain_sigmoid        = True,    # use logistic σ(·) (else raw sine)
        use_quadrature_feat = False,    # include quadratic + quadrature features
        seed                = 42       # RNG seed
    )



sph_res.fit_readout(train_input, train_target, discard=100)


trained_models = {
    "Cycle Res"      : cycle_res,
    "Sparse ESN"     : sparse_res,
    "Testing Res" : sph_res,
}


# 2c) Predict autoregressively 

initial_in   = test_input[0]
preds_all = {
    name: model.predict_autoregressive(initial_in, n_test_steps)
    for name, model in trained_models.items()
}

# ------------------------------------------------------------------
# (3) NRMSE table
# ------------------------------------------------------------------
horizons   = [200, 500, 1000]

nrmse_all = {
    name: evaluate_nrmse(preds, test_target, horizons)
    for name, preds in preds_all.items()
}

df_nrmse = pd.DataFrame(nrmse_all).T[horizons]
print("\n--- Multi-step NRMSE ---")
print(df_nrmse.to_string(float_format="%.5f"))

# ------------------------------------------------------------------
# (4) VPT table
# ------------------------------------------------------------------
lambda_max_lorenz = 0.9
threshold         = 0.4

vpt_all = {
    name: compute_valid_prediction_time(
              test_target, preds, time_test,
              threshold=threshold,
              lambda_max=lambda_max_lorenz
           )
    for name, preds in preds_all.items()
}

df_vpt = (pd.DataFrame(vpt_all, index=["T_VPT", "T_lambda", "T_VPT/T_lambda"])
            .T.round(3))

print(f"\n--- Valid Prediction Time (threshold={threshold}, "
      f"lambda_max={lambda_max_lorenz}) ---")
print(df_vpt.to_string())

# ------------------------------------------------------------------
# (5) ADev table
# ------------------------------------------------------------------
cube_size = (4, 4, 4)

adev_all = {
    name: compute_attractor_deviation(test_target, preds, cube_size)
    for name, preds in preds_all.items()
}

df_adev = pd.Series(adev_all, name="ADev").round(3)

print("\n--- Attractor Deviation ---")
print(df_adev.to_string())

```

```python
# ------------------------------------------------------------------
# 2-D line plot for every model in preds_all
# ------------------------------------------------------------------
plot_timeseries(
    time=time_test,                 # 1-D array of time stamps
    test_target=test_target,        # ground-truth trajectory
    preds_dict={"Testing Res": preds_all["Testing Res"]},       # preds_dict=preds_all,   for all   
    plot_len=3000,                  # truncate if desired
    dpi=300,                        # resolution
    save_as="lorenz_timeseries_comparison.png"   # saved in figures/
)

# ------------------------------------------------------------------
# 3-D trajectory plot for ONE model
# ------------------------------------------------------------------
plot_lorenz_3d(
    test_target=test_target,
    preds=preds_all["Testing Res"],
    plot_len=3000,
    dpi=300,
    save_as="lorenz_3d_testing_res.png"            # stored in figures/
)
```
